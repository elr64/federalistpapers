```python
import pickle
import argeparse
```

```python
def get_texts(generated_texts):
    with open(generated_texts, "r") as texts:
        hamilton=[]
        madison=[]
        madison_disputed=[]
        for line in texts:
            fields = line.strip().split("\t")
            author=fields[1]
            if author == 'hamilton_generated':
                hamilton.append(fields[2])
            if author == 'madison_generated':
                madison.append(fields[2])
            if author == 'madison1_generated':
                madison_disputed.append(fields[2])
    return hamilton, madison, madison_disputed
```
